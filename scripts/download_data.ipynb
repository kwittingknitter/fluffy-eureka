Download datasets from OLIS via the wrapper

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import wrapper

olis = wrapper.OlisAPI()
current_session = olis.get_current_session()
current_session
```

```python
sessions = olis.get_sessions()
sessions
```

```python
len(sessions['value'])
```

```python
import json

def set_default(obj):
    if isinstance(obj, set):
        return list(obj)
    return ''

os.makedirs('./../data', exist_ok=True) 

with open('./../data/current_session.json', 'w') as f:
    f.write(json.dumps(
    current_session['value'][0],
    indent=4,
    separators=(',', ': '),
    default=set_default
    ))

with open('./../data/sessions.json', 'w') as f:
    f.write(json.dumps(
    sessions['value'],
    indent=4,
    separators=(',', ': '),
    default=set_default
    ))
```
