Basic stats about bills from OR's last regular session, the 2025 regular session, by legislator

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
import wrapper

olis = wrapper.OlisAPI()
```

```python
current_session = olis.get_current_session()
current_legislators = current_session['value'][0]['Legislators']
len(current_legislators)
```

```python
session_key = "2025R1"
session_bills = olis.get_session_bills(session_key)
```

```python
session_bills['Measures']
```

```python
current_legislators[0]
```

```python
andersen_bills = olis.get_legislator_sponsored_bills(session_key, current_legislators[0]['LegislatorCode'])
andersen_chief_bills = olis.get_legislator_chief_sponsored_bills(session_key, current_legislators[0]['LegislatorCode'])
```

```python
andersen_bills['value']
```

```python
passed_bills = filter(lambda bill_info: bill_info['Measure']['CurrentLocation'] == 'Chapter Number Assigned', andersen_bills['value'])
len(list(passed_bills))
```

```python
passed_chief_bills = filter(lambda bill_info: bill_info['Measure']['CurrentLocation'] == 'Chapter Number Assigned', andersen_chief_bills['value'])
len(list(passed_chief_bills))
```

```python
reg_session_bill_count = len(session_bills['Measures'])

print('Total 2025 regular session bills:', reg_session_bill_count)
```

```python
for leg in current_legislators:
    leg_bills = olis.get_legislator_sponsored_bills(session_key, leg['LegislatorCode'])
    leg_chief_bills = olis.get_legislator_chief_sponsored_bills(session_key, leg['LegislatorCode'])
    # leg_passed_bills = olis.get_legislator_sponsored_bills_enrolled(session_key, leg['LegislatorCode'])
    leg_passed_bills = list(filter(lambda bill_info: bill_info['Measure']['CurrentLocation'] == 'Chapter Number Assigned', leg_bills['value']))

    print('{} {} {}: Sponsored {} bills, Chief sponsor for {} bills, Introduced {:.0%} of bills to session, Passed {:.0%} of bills sponsored'.format(
        leg['Title'], leg['FirstName'], leg['LastName'], 
        len(leg_bills['value']), len(leg_chief_bills['value']), 
        len(leg_bills['value'])/reg_session_bill_count, 
        len(leg_passed_bills)/len(leg_bills['value']))
    )
```
